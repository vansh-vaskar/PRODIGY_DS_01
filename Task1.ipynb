```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('population_clean.csv')

df_2020 = df[['Country Name', '2020']].copy()
df_2020 = df_2020.dropna(subset=['2020'])

plt.figure(figsize=(8, 5))
plt.hist(df_2020['2020'], bins=20)
plt.xlabel('Population in 2020')
plt.ylabel('Number of countries')
plt.title('Distribution of Total Population Across Countries in 2020')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('population_clean.csv')

df_2020 = df[['Country Name', '2020']].copy()
df_2020 = df_2020.dropna(subset=['2020'])

top10 = df_2020.sort_values('2020', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Country Name'], top10['2020'])
plt.xticks(rotation=45, ha='right')
plt.ylabel('Population in 2020')
plt.title('Top 10 Most Populated Countries in 2020')
plt.tight_layout()
plt.show()
```
